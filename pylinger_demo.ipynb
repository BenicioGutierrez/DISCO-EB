```python
import numpy as np
import matplotlib.pyplot as plt
from classy import Class # import classy module
from scipy.interpolate import interp1d
%matplotlib widget
```

```python
from pylinger_cosmo import cosmo
import pylinger_thermo as pthermo
from pylinger_pt import pt_synchronous
```

```python
## Cosmological Parameters
Tcmb = 2.7255
YHe = 0.248
Omegam = 0.276
Omegab = 0.0455
OmegaL = 1.0-Omegam
Omeganu = 0.0
num_massive_neutrinos = 0
mnu=0.0 #0.06 
Neff=3.046 #3.044
standard_neutrino_neff=Neff

h = 0.703
# mnu = 0.0
# mnu = 0.06
A_s = 2.1e-9
n_s = 0.965
k_p = 0.05
```

```python
## CLASS setup
# create instance of the class "Class"
LambdaCDM = Class()
# pass input parameters
LambdaCDM.set({
    'Omega_b':Omegab,
    'Omega_cdm':(Omegam-Omegab),
    'h':h,
    'A_s':A_s,
    'n_s':n_s,
    'N_ur':Neff,
    'tau_reio':0.06, 
    'T_cmb':Tcmb,
    'YHe':YHe,
    'output':'mPk,mTk,vTk',
    'lensing':'no',
    'P_k_max_1/Mpc':10.0,
    'z_max_pk':1000.0
  })
# run class
LambdaCDM.compute()
thermo = LambdaCDM.get_thermodynamics()

def get_class_power( fieldname, zout ):
  tk, k, z = LambdaCDM.get_transfer_and_k_and_z()
  TT = interp1d( np.log(1/(1+z)), tk[fieldname], axis=1 )( np.log(1/(1+zout)) ) 
  res =  2*np.pi**2*A_s*(k/k_p*h)**(n_s - 1) * k**(-3) * TT**2 *h**3 #/ (2*np.pi)**3 
  return res, k
```

```python
## pyLinGer setup
cp = cosmo(Omegam=Omegam, Omegab=Omegab, OmegaL=OmegaL, H0=100*h, Tcmb=Tcmb, YHe=YHe, Neff=Neff )
th = pthermo.thermal_history(taumin=cp.tau[1],taumax=cp.tau[-1],cp=cp,N=1000)
pt = pt_synchronous(cp=cp,th=th)

kmodes = np.geomspace(1e-3,1e1,100)
y,t = pt.compute(kmodes = kmodes, aexp_out=np.geomspace(1e-4,1e-2,2))
```

```python
## evaluate the power spectra
iout = 1
print(f'pyEB zout={1/y[0,0,iout]-1}')
zout = 1/y[0,0,iout]-1

Pkc = A_s*(kmodes/k_p)**(n_s - 1) * kmodes**(-3) * y[:,3,iout]**2 
Pkb = A_s*(kmodes/k_p)**(n_s - 1) * kmodes**(-3) * y[:,5,iout]**2 
Pkg = A_s*(kmodes/k_p)**(n_s - 1) * kmodes**(-3) * y[:,7,iout]**2 

Pkc_CAMB = PKc.P(zout,kmodes/h)
Pkb_CAMB = PKb.P(zout,kmodes/h)

Pkc_CLASS, k_CLASS = get_class_power('d_cdm', zout)
Pkb_CLASS, k_CLASS = get_class_power('d_b', zout)
```

```python
fig, ax = plt.subplots()
fac = 2.5
# fac = 2*np.pi**2/ (2*np.pi)**1.5 
ax.loglog( kmodes, Pkc*fac, label='pyEB cdm', color='C0')
# ax.loglog( kmodes, Pkc_CAMB, label='CAMB cdm', color='C0', ls='--')
ax.loglog( kmodes, Pkb*fac, label='pyEB baryon', color='C1') 
# ax.loglog( kmodes, Pkb_CAMB, label='CAMB baryon', color='C1', ls='--')
ax.loglog( k_CLASS, Pkc_CLASS, label='CLASS cdm', ls=':', color='C0')
ax.loglog( k_CLASS, Pkb_CLASS, label='CLASS baryon', ls=':', color='C1')
ax.legend()
# ax.loglog( kmodes, Pkg, label='pyEB photon', color='C2') 
# ax.set_xlim((1e-3,5))
```

```python
def nu2( a : float, amnu : float, lmaxnu : int, psi0, psi1, psi2 ):
    const = 7/120 * math.pi**4 
    qmax = 25.0
    # MAXJ=3 stops Richardson extrapolation at three levels, so that the
    # integration scheme is 8th-order Newton-Cotes.
    MAXITER = 7
    MAXJ = 3

    g0 = np.zeros((4))
    gmax = np.zeros((4))
    g = np.zeros((4,MAXJ+1))
    
    if nqmax==0:
      drhonu=0.0
      fnu=0.0
      dpnu=0.0
      shearnu=0.0
      return drhonu,fnu,dpnu,shearnu
    
    # Compute the integrals
    nqmax = 2**MAXITER
    # q is the comoving momentum in units of k_B*T_nu0/c.
    # Integrate up to qmax and then use asymptotic expansion for remainder.
    h=qmax/2
    q=qmax
    aq=a*amnu/q
    v=1.0/jnp.sqrt(1.0+aq*aq)
    qdn=h*q*q*q/(jnp.exp(q)+1.0)
    gmax = jnp.array([qdn*psi0[-1]/v, qdn*psi0[-1]*v, qdn*psi1[-1], qdn*psi2[-1]*v])
    g[:,0] = gmax
    gf = jnp.copy(gmax)
    nint=1
    nn=nqmax//2
    for iter in range(MAXITER):
      # Calculate next trapezoidal rule approximation to integral.
      g0 = jnp.zeros((4))
      for k in range(nint):
        q=(k+k+1)*h
        iq=(k+k+1)*nn
        aq=a*amnu/q
        v=1.0/jnp.sqrt(1.0+aq**2)
        qdn=h*q**3/(jnp.exp(q)+1.0)
        g0.at[0].set( g0[0]+qdn*psi0[iq]/v )
        g0.at[1].set( g0[1]+qdn*psi0[iq]*v )
        g0.at[2].set( g0[2]+qdn*psi1[iq] )
        g0.at[3].set( g0[3]+qdn*psi2[iq]*v )
	
      g0 = 0.5 * g[:,0] + g0
      h = 0.5 * h
      nint = nint + nint
      nn = nn // 2
      jmax = min(iter,MAXJ)
      fourj = 1.0
      for j in range(jmax):
        # Use Richardson extrapolation.
        fourj = 4.0 * fourj
        newg0 = g0 + (g0 - g[:,j])/(fourj-1.0)
        g[:,j] = g0
        g0 = newg0
      err = 1.0-gmax/g0
      gmax = g0
      g[:,jmax] = g0

    drhonu = (g0[0] + gf[0] * 2 / gmax) / const
    dpnu = (g0[1] + gf[1] * 2 / gmax) / const / 3.0
    fnu = (g0[2] + gf[2] * 2 / gmax) / const
    shearnu = (g0[3] + gf[3] * 2 / gmax) / const * 2.0/3.0

    return drhonu,fnu,dpnu,shearnu
```
