```python
def get_sigma_M_( *, M, Pkm, kmodes, param ):
    """ compute the variance of the density field smoothed on a scale R
    
    Args:
        Pkm (array_like) : the power spectrum
        kmodes (array_like) : the wavenumbers
        M (array_like) : the mass
        param (dict) : dictionary of all parameters
        
    Returns:
        sigma (array_like) : the variance
    """
    rho_crit = 1e-4*param['H0']**2 * 2.77440334e11 # Msun / Mpc
    Wth = lambda kR : 3*(jnp.sin(kR)-kR*jnp.cos(kR))/(kR)**3
    R = (3*M/(4*jnp.pi*param['Omegam']*rho_crit))**(1/3)


    dsigma = Wth(kmodes*R)**2 * Pkm * kmodes**2
    return jnp.sqrt(jnp.trapz( y=(kmodes*dsigma), x=jnp.log(kmodes) ) / (2*jnp.pi))

def get_sigma_M( *, M, Pkm, kmodes, param ):
    return jax.vmap( lambda mm: get_sigma_M_(M=mm, Pkm=Pkm, kmodes=kmodes, param=param) )(M)
```

```python
def get_Tinker_coeffs( Delta ):
  vDelta = jnp.array([200, 300, 400, 600, 800, 1200, 1600, 2400, 3200])
  vA = jnp.array([0.186, 0.200, 0.212, 0.218, 0.248, 0.255, 0.260, 0.260, 0.260])
  va = jnp.array([1.47, 1.52, 1.56, 1.61, 1.87, 2.13, 2.30, 2.53, 2.66])
  vb = jnp.array([2.57, 2.25, 2.05, 1.87, 1.59, 1.51, 1.46, 1.44, 1.41])
  vc = jnp.array([1.19, 1.27, 1.34, 1.45, 1.58, 1.80, 1.97, 2.24, 2.44])

  A = jnp.interp( jnp.log10(Delta), jnp.log10(vDelta), vA )
  a = jnp.interp( jnp.log10(Delta), jnp.log10(vDelta), va )
  b = jnp.interp( jnp.log10(Delta), jnp.log10(vDelta), vb )
  c = jnp.interp( jnp.log10(Delta), jnp.log10(vDelta), vc )

  return A, a, b, c
```

```python
def f_SC( a, y, args ):
  param = args[0]
  dy = jnp.zeros((4,))
  
  # ... quintessence
  def E_( a, param ):
      # massive neutrinos
      rhonu = jnp.exp(param['logrhonu_of_loga_spline'].evaluate(jnp.log(a)))
    #   pnu = jnp.exp(param['logpnu_of_loga_spline'].evaluate( jnp.log(a) ) )

    #   w_Q                = param['w_DE_0'] + param['w_DE_a'] * (1.0 - a) 
      rho_Q              = a**(-3*(1+param['w_DE_0']+param['w_DE_a'])) * jnp.exp(3*(a-1)*param['w_DE_a'])
  
      # ... homogeneous background
      grho = (
          param['grhom'] * param['Omegam'] / a
          + (param['grhog'] + param['grhor'] * (param['Neff'] + param['Nmnu'] * rhonu)) / a**2
          + param['grhom'] * param['OmegaDE'] * rho_Q * a**2
      )

      return jnp.sqrt(grho / param['grhom'] )/a                # Friedmann I
  
  E      = E_( a, param )
  Eprime = jax.grad( lambda aa: E_( aa, param) )(a)

  dy = dy.at[0].set( y[1] )
  dy = dy.at[1].set( -(3/a + Eprime/E) * y[1] + 4/3 * y[1]**2 / (1+y[0]) \
          + 3/2 * y[0] * (1+y[0]) * param['Omegam'] / a**5 / E**2 )

  dy = dy.at[2].set( y[3] )
  dy = dy.at[3].set( -(3/a + Eprime/E) * y[3] + 3/2 * y[2] * param['Omegam'] / a**5 / E**2 )


  # ## EdS:
  # dy = dy.at[0].set( y[1] )
  # dy = dy.at[1].set( -(3/a + Eprime/E) * y[1] + 4/3 * y[1]**2 / (1+y[0]) \
  #         + 3/2 * y[0] * (1+y[0]) / a**5 / E**2 )

  # dy = dy.at[2].set( y[3] )
  # dy = dy.at[3].set( -(3/a + Eprime/E) * y[3] + 3/2 * y[2] / a**5 / E**2 )

  return dy
```

```python

# @jax.jit
def get_collapse_threshold(param):
  amin = 1e-3
  amax = 1e1
  
  term = diffrax.ODETerm(f_SC)
  solver = diffrax.Dopri5()
  stepsize_controller = diffrax.PIDController(rtol=1e-4, atol=1e-4)
  saveat = diffrax.SaveAt(dense = True)
  
  def cond_collapse( logd0, args ):
      param, amin, amax, term, solver, saveat, stepsize_controller = args
      d0 = jnp.exp(logd0)
      y0 = jnp.array([d0 * amin, d0, d0*amin, d0])
      solution = diffrax.diffeqsolve(term, solver, t0=amin, t1=amax, dt0=amin, y0=y0, args=(param,), 
                                     saveat=saveat, stepsize_controller=stepsize_controller, throw=False,
                                     adjoint=diffrax.DirectAdjoint() )
      # aa = zip(solution.interpolation.ts, solution.interpolation.infos['y1'][:,0])
      
      aa = solution.interpolation.ts
      amax = jnp.where( jnp.isfinite(aa), aa, 0.0 ).max()

      # dnlin   = solution.interpolation.infos['y1'][:,0]
      # acoll = solution.interpolation.ts[1+jnp.argmax(dnlin > 1e8)]  # returns amin if no collapse
      acoll = amax
      return acoll - 1.0
      
  logd0c = root_find_bisect(func=cond_collapse, xleft=-1.0, xright=1.0, numit=5, param=(param,amin,amax,term,solver,saveat,stepsize_controller) )
  d0c = jnp.exp(logd0c)
  
  y0 = jnp.array([d0c * amin, d0c, d0c*amin, d0c])
  solution = diffrax.diffeqsolve(term, solver, t0=amin, t1=amax, dt0=amin, y0=y0, args=(param,), saveat=saveat, 
                                 stepsize_controller=stepsize_controller, throw=False, adjoint=diffrax.DirectAdjoint() )
      
  dlin   = solution.interpolation.infos['y1'][:,2]

  deltac = jnp.where( jnp.isfinite(dlin), dlin, 0.0 ).max()
  
  return deltac
```

```python
def compute_MF( theta ):
  param = {}
  param['Omegam']  = theta[0]
  param['Omegab']  = theta[1]
  param['OmegaDE'] = 1 - param['Omegam']
  param['w_DE_0']  = theta[2]
  param['w_DE_a']  = theta[3]-1
  param['cs2_DE']  = 1.0
  param['Omegak']  = 0.0
  param['A_s']     = 2.1064e-09
  param['n_s']     = theta[4]
  param['H0']      = 100*theta[5]
  h = theta[5]
  param['Tcmb']    = 2.7255
  param['YHe']     = 0.248
  param['Neff']    = 2.046
  param['Nmnu']    = 1.0
  param['mnu']     = theta[6]
  param['k_p']     = 0.05
  param['sigma_8'] = theta[7]

  # compute background evolution and thermal history
  param = evolve_background(param=param, thermo_module='RECFAST')

  # compute perturbations at aexp = theta[8]
  nmodes = 512
  kmin = 1e-3
  kmax = 1e+2

  aexp_out = jnp.array([theta[8], 1.0])
  y, kmodes = evolve_perturbations( param=param, kmin=kmin, kmax=kmax, num_k=nmodes, aexp_out=aexp_out )

  # get power spectrum, normalised to sigma_8
  Pkm0 = get_power( k=kmodes, y=y[:,-1,:], idx=4, param=param )
  Pkm  = get_power( k=kmodes, y=y[:,0,:], idx=4, param=param )

  Wth = lambda kR : 3*(jnp.sin(kR)-kR*jnp.cos(kR))/(kR)**3
  dsigma0 = Wth(kmodes*8.0/h)**2 * Pkm0 * kmodes**2

  sigma8_2_measured = jnp.trapz( y=(kmodes*dsigma0), x=jnp.log(kmodes) ) / (2*jnp.pi)

  Pkm  = Pkm  * param['sigma_8']**2 / sigma8_2_measured

  # compute mass function based on various prescriptions
  Delta = 200 / param['Omegam']    # we want to use 200 times the critical density
  Mmin  = 1e13
  Mmax  = 1e16
  nM    = 512

  rho_crit = 1e-4*param['H0']**2 * 2.77440334e11 # Msun / Mpc
  rhobar = param['Omegam']*rho_crit

  M = jnp.geomspace( Mmin, Mmax, nM )
  sigmaM = get_sigma_M( Pkm=Pkm, kmodes=kmodes, M=M, param=param )

  dinvlogsigma_dM_func = jax.grad( lambda mm: jnp.log(1/get_sigma_M_( Pkm=Pkm, kmodes=kmodes, M=mm, param=param)) )
  dinvlogsigma_dM = jax.vmap( dinvlogsigma_dM_func )(M)

  mf_type = 'mPS'

  if mf_type == 'Tinker':
    # Tinker et al. 2008 mass function fit
    deltac = 1.686
    A, a, b, c = get_Tinker_coeffs( Delta )
    f = A * ((sigmaM/b)**(-a) + 1) * jnp.exp(-c/sigmaM**2)
    
  elif mf_type == 'PS':
    # Press-Schechter mass function
    deltac = get_collapse_threshold( param )
    fac = jnp.sqrt(2/jnp.pi) * deltac
    f = 1/2 * fac / sigmaM * jnp.exp( -deltac**2 / (2*sigmaM**2) )

  elif mf_type == 'mPS':
    #modified Press-Schechter mass function from Sheth-Tormen
    deltac = get_collapse_threshold( param )
    nup = 0.707 * (deltac/sigmaM)**2
    p = 0.3
    A = 0.455705
    f = A * (1 + (1/nup)**p) * jnp.sqrt(nup/2/jnp.pi) * jnp.exp(-nup/2)

  elif mf_type == 'Zeldo':
    deltac = get_collapse_threshold( param )
    lambdac_o_deltac = 0.22539
    nup = (lambdac_o_deltac)**2 * (deltac/sigmaM)**2

    Za = 25/4 * jnp.sqrt(10*nup/jnp.pi)*(5*nup/3-1/12)*jnp.exp(-5*nup/2)*jax.scipy.special.erfc(jnp.sqrt(2*nup))
    Zb = 25/16 * jnp.sqrt(15*nup/jnp.pi)*jnp.exp(-15*nup/4)*jax.scipy.special.erfc(jnp.sqrt(3*nup/4))
    Zc = -125/12 * jnp.sqrt(5)/jnp.pi * nup * jnp.exp(-9*nup/2)

    f = Za + Zb + Zc

  dndM = f * rhobar / M * dinvlogsigma_dM 

  return dndM*M / h**3, M * h, deltac
```
